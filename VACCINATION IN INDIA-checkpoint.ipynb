```python
# NOW WE ARE USING LINEAR REGRESSION TO FIND  TOTAL PEOPLE OF DAILY VACCINATION
```

```python
data
```

```python
x=data.iloc[:,2:7].values
y=data.iloc[:,7].values
```

```python
y
```

```python
x
```

```python
# Import the model and traing and splitting
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(x, y, test_size=0.3, random_state=10)
```

```python
# Fit the model
linreg.fit(X_train, y_train)
```

```python
# Predict the value FOR PEOPLE VACCINATED AT RESPECTIVE DAY
my_pred = linreg.predict(X_test)
my_pred 
```

```python
# Calculate the accuracy score
my_score = linreg.score(X_test, y_test)
my_score
```
